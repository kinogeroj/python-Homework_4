### Задание 1.1.
Скачать данные.

```python
import matplotlib.pyplot as plt

import pandas as pd

import seaborn as sns

df = pd.read_csv('./kc_house_data.csv')

df.head()
```

### Задание 1.2.
Изучите стоимости недвижимости.

```python
plt.figure(figsize = (6, 4))

plt.hist(df['price'], bins=30)

plt.title('Анализ стоимости недвижимости')

plt.xlabel('Стоимость, у.е.')

plt.ylabel('Количество домов')

print('Выводы:\n 1. Большая часть домов стоит в районе 540-550 тысяч у.е.\n 2. Наибольшее число домов продаются за большую стоимость.')
```

### Задание 1.3.
Изучите распределение квадратуры жилой площади.

```python
plt.figure(figsize = (6, 4))

plt.hist(df['sqft_living'], bins = 30)

plt.title('Распределение квадратуры жилой площади')

plt.xlabel('Жилая площадь, кв. футы')

plt.ylabel('Количество домов')

print('Выводы:\n 1. Наибольшую стоимость имеют дома с площадью в диапазоне от 1000 до 3000 кв.м.\n 2. Наибольшее кол-во домов имеют площадь менее 3000 кв. футов.')
```

### Задание 1.4.
Изучите распределение года постройки.

```python
plt.figure(figsize = (6, 4))

plt.hist(df['yr_built'], bins = 30)

plt.title('Распределение годов постройки')

plt.xlabel('Год постройки')

plt.ylabel('Количество домов')

print('Выводы:\n 1. Строительство домов началось в 1900 году.\n 2. С течением времени скорость постройки возрастает.\n 3. Спад в темпах строительства новых домов наблюдался в 1930-х годах и 1970-х гг.\n 4. Пик строительства приходится на начало 2000-х гг.')
```

## Задание 2.

### Задание 2.1.
Изучите распределение домов от наличия вида на набережную.
Постройте график.
Сделайте выводы.

```python
waterfront_distribution = df['waterfront'].value_counts()

plt.figure(figsize = (8, 8))

plt.pie(waterfront_distribution.values, autopct = '%1.1f%%')

plt.title('Распределение кол-ва домов, в зависимости от наличия вида на набережную')

plt.legend(['Без вида на набережную', 'С видом на набережную'])

print('Вывод: кол-во домов с видом на набережную составляет менее 1%, что создает острый дефицит.')
```

### Задание 2.2.
Изучите распределение этажей домов.

```python
floors_distribution = df['floors'].value_counts()

plt.figure(figsize = (8, 8))

plt.pie(floors_distribution.values, autopct='%1.1f%%')

plt.title('Распределение этажей домов')

plt.legend(floors_distribution.index)

print('Выводы:\n 1. Половина домов являются одноэтажными.\n 2. Почти 90% домов являются одно- или двух-этажными.')
```

### Задание 2.3.
Изучите распределение состояния домов.

```python
condition_distribution = df['condition'].value_counts()

plt.figure(figsize = (8, 8))

plt.pie(condition_distribution.values, autopct='%1.1f%%')

plt.title('Распределение состояния домов')

plt.legend(condition_distribution.index)

print('Вывод: 90% недвижимости находятся в хорошем и удовлетворительном состоянии.')
```

## Задание 3. 
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

Возьмем в качестве ключевых параметров общую площадь, жилую площадь, этажность, состояние дома и год постройки. Выясним, как будет меняться цена в зависимости от этих характеристик.

```python
plt.figure(figsize = (8, 5))

plt.scatter(df['price'], df['sqft_lot'])

plt.title('Зависимость стоимости дома от общей площади')

plt.xlabel('Стоимость, у.е.')

plt.ylabel('Общая площадь дома, кв. футы')

print('График построен на основе имеющейся таблицы данных')
```

Из графика можно сделать вывод, что цена не имеет видимой линейной зависимости от общей площади дома, т.е. с ростом общей площади цена необязательно увеличивается. Дома с небольшой общей площадью могут иметь высокую стоимость, в то время как дома с максимальной общей площадью имеют стоимость ниже. Очевидно, общая площадь дома не является фактором, влияющим на цену напрямую.

```python
plt.figure(figsize = (8, 5))

plt.scatter(df['price'], df['sqft_living'])

plt.title('Зависимость стоимости дома от жилой площади')

plt.xlabel('Стоимость, у.е.')

plt.ylabel('Жилая площадь дома, кв. футы')

print('График построен на основе имеющейся таблицы данных')
```

Очевидно, что, в целом, существует зависимость стоимости дома от его жилой площади. Причем, чем больше жилая площадь, тем выше стоимость дома.

```python
plt.figure(figsize = (8, 5))

plt.scatter(df['price'], df['floors'])

plt.title('Зависимость стоимости дома от кол-ва этажей')

plt.xlabel('Стоимость, у.е.')

plt.ylabel('Кол-во этажей')

print('График построен на основе имеющейся таблицы данных')
```

Линейная зависимость цены четко прослеживается для каждой категории домов с определенной этажностью и зависит от других характеристик домов, влияющих на ценообразование. Явной зависимости дома от его этажности, в целом, не прослеживается, однако, наиболее дорогими являются двух-этажные дома.

```python
plt.figure(figsize = (8, 5))

plt.scatter(df['price'], df['condition'])

plt.title('Зависимость цены дома от его состояния')

plt.xlabel('Стоимость, у.е.')

plt.ylabel('Состояние дома')

print('График построен на основе имеющейся таблицы данных')
```

График показывает, что, чем лучше состояние дома, тем он дороже. Но, как и в предыдущем случае, линейная зависимость цены по большей мере прослеживается именно в конкретных категориях домов.

```python
plt.figure(figsize=(8, 5))

plt.scatter(df['price'], df['yr_built'])

plt.title('Зависимость стоимости дома от года постройки')

plt.xlabel('Стоимость, у.е.')

plt.ylabel('Год постройки')

print('График построен на основе имеющейся таблицы данных')
```

Анализируя представленный график, можно сделать вывод, что год постройки слабо влияет на ценообразование дома.

**Вывод.**
Из выбранных нами параметров, наибольшее влияние на стоимость дома оказывает именно его жилая площадь. Остальные параметры играют менее существенную роль в ценообразовании и явной линейной зависимости не прослеживается. В подтверждение данного вывода, ниже проведем корреляционный анализ.

Создадим отдельную таблицу с выбранными нами ранее параметрами.

```python
conclusion = df[['price', 'sqft_lot', 'sqft_living', 'yr_built', 'floors', 'condition']]

conclusion.head(10)
```

Построим тепловую карту для определения линейных связей выбранных нами параметров.

```python
sns.heatmap(conclusion.corr(), annot = True, linewidths = 1.5, cmap = 'coolwarm')

print('График построен на основе имеющейся таблицы данных')
```

Из тепловой карты мы видим, что, как и предполагалось, явной линейной зависимости стоимости дома от выбранных факторов не наблюдается, за исключением жилой площади, связь с которой явно видна.
